## SENTIMENT ANALYSIS=================

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    Trainer,
    TrainingArguments,
    DataCollatorWithPadding
)
import evaluate
import torch
```

```python
print(review_df.shape) 
```

- 7 M rows are big, lets try to sample 20 % of them first

```python
df = review_df[['text', 'stars']].dropna().sample(frac=0.2, random_state=42)  
df['label'] = (df['stars']-1).astype(int)
```

```python
df
```

##  train test split 0.9 training and 0.1 for evaluation

```python
train_df, eval_df = train_test_split(df, test_size=0.1, stratify=df['label'], random_state=42) 
train_dataset = Dataset.from_pandas(train_df)
eval_dataset   = Dataset.from_pandas(eval_df)
```

# also define what will be test set

```python
test_df= review_df.drop(df.index).sample(n=len(df), random_state=123)
test_df['label'] = (test_df['stars']-1).astype(int)
print(f"train_df rows: {len(train_df)}")
print(f"eval_df rows: {len(eval_df)}")
print(f"df rows: {len(df)}")
print(f"test_df rows: {len(test_df)}")
# make sure no overlap between train/eval and test sets
print(f"Overlap:  {len(set(df.index) & set(test_df.index))}")
```

## Choose Model

```python
model_name = "microsoft/deberta-v3-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
def preprocess(examples):
    return tokenizer(examples["text"], truncation=True, max_length=256)  # 512 safe, 256 faster

tokenized_train = train_dataset.map(preprocess, batched=True)
tokenized_eval  = eval_dataset.map(preprocess, batched=True)
```

```python
tokenized_train = tokenized_train.remove_columns(["text", "stars"])
tokenized_eval  = tokenized_eval.remove_columns(["text", "stars"])
tokenized_train.set_format("torch")
tokenized_eval.set_format("torch")
```

```python
print(tokenized_train)
print(tokenized_eval)
```

```python
model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=5,
    torch_dtype=torch.float32,  # ← force fp32
    device_map="auto"           # optional, but good on single GPU
)
```

## Metrics

```python
accuracy = evaluate.load("accuracy")
mae_metric = evaluate.load("mae")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    acc = accuracy.compute(predictions=preds, references=labels)["accuracy"]
    mae = mae_metric.compute(predictions=preds, references=labels)["mae"]  # 0–4 scale
    return {"accuracy": acc, "mae": mae}  # MAE ~0.6 is good
```

## TRAIN

```python
import torch
import gc

torch.cuda.empty_cache()
gc.collect()
torch.cuda.synchronize()

training_args = TrainingArguments(
    output_dir="./yelp_deberta_sentiment",
    num_train_epochs=3,                  # 3–4 usually enough
    per_device_train_batch_size=8,    
    per_device_eval_batch_size=16,
   gradient_accumulation_steps=4,       # effective batch 16 if bs=8
    learning_rate=2e-5,                  # standard sweet spot
    weight_decay=0.01,
    warmup_ratio=0.1,
    fp16=True,                      # ← turn off AMP
    bf16=False,                        # only if you have Ampere+ and want to try
    eval_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    metric_for_best_model="accuracy",    # or "mae" with greater_is_better=False
    greater_is_better=True,
    logging_steps=200,
    report_to="none",                    # or "tensorboard" if you like
                 
gradient_checkpointing=True,     # huge memory saver for transformers
)


trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_eval,
    data_collator=DataCollatorWithPadding(tokenizer=tokenizer),
    compute_metrics=compute_metrics,
)
trainer.train()
trainer.save_model("./yelp_sentiment_model")
tokenizer.save_pretrained("./yelp_sentiment_model")
```

```python
from transformers import pipeline

pipe = pipeline("text-classification", model="./yelp_sentiment_model", tokenizer=tokenizer, device=0)

examples = [
    "Absolutely loved the dim sum here — fresh, flavorful, and great value!",
    "Food was cold, service rude, and place dirty. Never coming back.",
    "It was average, nothing special but decent for the price."
]

for ex in examples:
    result = pipe(ex)[0]
    pred_label = int(result['label'].split('_')[-1]) + 1  # LABEL_0 → 1 star, etc.
    print(f"Review: {ex}\nPredicted: {pred_label} stars (conf: {result['score']:.3f})\n")
```

# full set evaluation

```python
import pandas as pd
import numpy as np
from transformers import pipeline
from sklearn.metrics import accuracy_score, classification_report, mean_absolute_error

# ────────────────────────────────────────────────
# 1. Load the fine-tuned pipeline
# ────────────────────────────────────────────────
pipe = pipeline(
    "text-classification",
    model="./yelp_sentiment_model",
    tokenizer="./yelp_sentiment_model",  # or your original tokenizer object
    device=0,                            # GPU
    batch_size=32,                       # adjust based on VRAM (16–64 usually good)
    truncation=True,
    max_length=512,
)

# ────────────────────────────────────────────────
# 2. Predict on the entire DataFrame (in batches)
# ────────────────────────────────────────────────
print("Running predictions on all rows...")

# Get predictions (returns list of dicts: [{'label': 'LABEL_X', 'score': 0.XX}, ...])
results = pipe(df['text'].tolist())

# Extract predicted label index (0–4) and confidence
pred_labels = []   # 0–4
pred_scores = []   # confidence

for res in results:
    # 'LABEL_0' → 0, 'LABEL_1' → 1, etc.
    label_str = res['label']
    label_idx = int(label_str.split('_')[-1])
    pred_labels.append(label_idx)
    pred_scores.append(res['score'])

# Convert to predicted stars (1–5)
df['pred_label'] = pred_labels          # 0–4
df['pred_stars'] = df['pred_label'] + 1 # 1–5
df['pred_confidence'] = pred_scores

print(f"Predictions completed. Shape: {df.shape}")

df.head()
```

```python

y_true = (df['stars'] - 1).round().astype(int).values
y_pred = df['pred_label'].values

# Classification metrics (5-class)
accuracy = accuracy_score(y_true, y_pred)
print("\nAccuracy (exact match 1–5 stars): {:.4f} ({:.2f}%)".format(accuracy, accuracy * 100))


print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=['1★', '2★', '3★', '4★', '5★']))

# Regression-style metrics (useful for star prediction)
mae = mean_absolute_error(y_true + 1, y_pred + 1)   # in star scale (1–5)
rmse = np.sqrt(np.mean((y_true + 1 - (y_pred + 1))**2))

print(f"\nMean Absolute Error (MAE)  : {mae:.3f} stars")
print(f"Root Mean Squared Error (RMSE): {rmse:.3f} stars")
```
